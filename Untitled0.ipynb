```python
import matplotlib.pyplot as plt
```

```python
x=[i for i in range(10)]
print(x)
```

```python
y=[2*i for i in range (10) ]
print(y)
```

```python
plt.plot(x,y)
```

```python
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
```
